```python
def colors(states):
  color_list= []
  for state in states:
   if state == 'Completed':
    color_list.append('darkslategrey')
   elif state == 'Shopping basket':
    color_list.append('silver')
   elif state == 'Cancelled':
    color_list.append('lightgrey')
   else:
    color_list.append('gainsboro')
  return color_list
```
